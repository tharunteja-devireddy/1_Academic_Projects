<a id = 'data'></a>
### Data
- we will be using randomly generated data by using sklearn clusters function
- in order to enable visualising our generated data will consist of only 2 dimentional 
  features
---------------------------------------------------------------------------------- [BACK TO TOP](#top)

<a id = 'load'></a>
#### loading the data

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=1000,n_features=2,centers=4,cluster_std=2) 
# creating 1000 samples with 2 features each and 4 cluster centers and with points far away from the cluster
# center as cluster_std is set higher
xdata = data[0] # features 
ydata = data[1] # cluster to which this feature belongs to 
print(xdata.shape,ydata.shape)
```

<a id = 'analyse'></a>
#### visualizing the data
- the below plot represents how our actual data needs to be clustered let us see
how our models perform on this data later in the below sections

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=xdata[:,0],y=xdata[:,1],hue=ydata,palette='rainbow')
plt.title('Generated Data')
plt.grid()
```

<a id = 'kmeans'></a>
### clustering using kmeans
---------------------------------------------------------------------------------- [BACK TO TOP](#top)

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=4) # mentioning the no of clusters into which we need to cluster our data
kmean.fit(xdata)     # clustering our data
yhat1 = kmean.labels_ # extracting the cluster of each datapoint

plt.figure(figsize=(10,6))
sns.scatterplot(x=xdata[:,0],y=xdata[:,1],hue=yhat1,palette='rainbow')
plt.title('Clustering using K-means')
plt.grid()
```

<a id = 'dbscan'></a>
### clustering using dbscan
---------------------------------------------------------------------------------- [BACK TO TOP](#top)

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1,min_samples=10) # 
dbscan.fit(xdata)
yhat2 = dbscan.labels_

plt.figure(figsize=(10,6))
sns.scatterplot(x=xdata[:,0],y=xdata[:,1],hue=yhat2,palette='rainbow')
plt.title('Clustering using K-means')
plt.grid()
```

<a id = 'conclusion'></a>
### conclusion
- dbscan being density based clustering failed to differentiate between 3rd and 4th
clusters due to their proximity to each other 
- but dbscan has enabled us to differentiate outliers
- kmeans has clustered the data into 4 clusters exactly as our generated data
  kmeans has no concept of ouliers so it failed to identify them
